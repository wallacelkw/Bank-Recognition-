**Image Processing Task:**
1. Create An Array
2. Create A Loop Function
2. Read The Image File
3. Resizing Image
4. Converting Image To Gray Scale
5. Flattening Image
6. Storing into Array

```python
image_array = []
for img_file in listdir(paths):
    img = cv.imread(paths+img_file)
    print(img)
    img_resized = cv.resize(img, (400, 200))
    img_gray = cv.cvtColor(img_resized, cv.COLOR_BGR2GRAY)
    img_1D_vector = img_gray.flatten().reshape(-1)
    image_array.append(img_1D_vector)
    print(img_1D_vector)
```
