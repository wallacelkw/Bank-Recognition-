```python
import glob
import cv2
import os
from pathlib import Path
```

```python
img_paths = 'All_images'
```

```python
images_data = []

for img in glob.glob(img_paths + "/*.jpg"):
    filename = Path(img).stem
    images_data.append(filename)
```

```python
print(images_data)
```

```python
print(len(images_data))
```

```python
import re
store_data =[]
for text in images_data:
    result = re.sub('_\d+', '', text, flags=re.IGNORECASE)
    store_data.append(result)

store_data
```

```python
print(len(store_data))
```

```python
import pandas as pd
# Store the label into csv file
fields = ['Label']
data = {'Label' : store_data}
df = pd.DataFrame(data)
df.to_csv('G6_Label.csv',index=False)
```
